```python
import numpy as np    #计算数据
import matplotlib.pyplot as plt  #绘制图线
import pandas as pd
```

```python
def getOriData():
    #读取原始CSV数据 未经处理
    train = pd.read_csv('mnist_train.csv',header=None)  #没有表头 header指定为None
    test = pd.read_csv('mnist_test.csv',header = None)
    return train,test
```

```python
#二值化映射函数
def Binarization(val):
    tag = 255/2
    if val < tag:
        return 0
    else:
        return 1
```

```python
#原始训练数据集、测试数据集
trainData,testData = getOriData()
trainData.shape[0]
```

```python
for i in range(trainData.shape[0]):
    #二维数组的切片不易理解 转换为一维切片更易理解
    trainData.loc[i][1:]=trainData.loc[i][1:].apply(Binarization)
```

```python
#拆分标签和特征向量
label = trainData.iloc[:,0]
feature = trainData.iloc[:,1:]
```

```python
def getPCJ(data):
    #计算先验概率 P(C|J)
    tag=0
    counts = [0]*10
    for i in range(data.shape[0]):
        tag=data.loc[i][0]
        counts[tag]+=1
    PCJ=np.zeros(10)  #先验概率
    for i in range(10):
        PCJ[i] = counts[i]/60000
    return counts,PCJ
```
